```r
library(dplyr)
library(ggplot2)

mydata <- read.csv('https://academics.hamilton.edu/mathematics/ckuruwit/Data/breastcancer2.csv',
                    stringsAsFactors = TRUE)
                    # use stringsAsFactors to properly read character variables into R
```

```r
#convert status to 0-1 var
mydata$cancercode <- recode(mydata$status, "M" = 1, "B" = 0)
```

```r
# Suppose we are interested in building a model with the
# radius variabele.  Explore how this variable associated with malignancy

mydata %>%
ggplot() +
  aes( x = radius, y = cancercode) +
  geom_jitter(height = 0.05, width = 0.01, alpha = 0.5, aes(col = status))+
  geom_smooth(method = "glm", method.args = list(family = "binomial"), se = FALSE)
```

```r

mymodel <- glm(cancercode ~ radius, family = binomial, data = mydata)
#tells r studio the y value belongs to a binomial distribution
summary(mymodel)
# equation = -15.2 + 1.03(radius) is wrong (that is linear model)
# should be e^(-15.2 + 1.03(radius))/1 + e^(-15.2 + 1.03(radius))
```


special latex notation:
$\hat p = \frac{e^-15.2 + 1.03(radius)}{1 + e^-15.2 + 1.03(radius)}$

where $\hat p $ is the predicted prob of cancer using this model

how to interpret the slope?

first acknowledge the sign, since coefficient is positive overally the increase in radius results in a correlate increase in predicted probability of cancer.


```r
mynewdata <- data.frame (radius = c(12,13,17,18))
predicted_prob <- predict(mymodel, newdata = mynewdata, type = "response")
head(predicted_prob)
```
